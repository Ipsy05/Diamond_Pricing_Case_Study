### Investigating the Variation of Diamond Prices

In this analysis, the primary focus is on investigating the variation in prices of diamonds without considering their relationship with any other variable. Studying a single variable, in this case, the price, can provide valuable insights into the overall distribution and characteristics of diamond prices.

#### Observations and Considerations:

1. **Highest Price Investigation:**
   - One can easily determine the highest price of a diamond through this analysis.

2. **Distribution of Prices:**
   - By studying only the diamond prices, we can gain insights into the distribution of prices across the dataset.

#### Utilizing Histograms for Visualization:

- **Histograms:**
  - To showcase the distribution of a variable, especially in the case of diamond prices, histograms prove to be an effective visualization tool.
  - Histograms allow the creation of 'bins' in the variable (in this case, prices) and plot the count of data points in each bin.
  - The resulting histogram provides a clear representation of how prices are distributed, highlighting patterns and trends in the dataset.

By focusing on the variation of diamond prices through histograms, we aim to uncover essential characteristics of the dataset, offering a foundational understanding of the distribution of prices.


```python
# Read the data into a dataframe
data = pd.read_csv("diamonds.csv", header=0)

# As a preparatory step, let's drop the 'Unnamed' column.
data = data.drop("Unnamed: 0", axis=1)

# Investigate the distribution of diamond prices using a histogram
plt.figure(figsize=(10, 6))
plt.hist(data['price'], bins=100, color='skyblue', edgecolor='black')
plt.title('Distribution of Diamond Prices', fontdict={'fontsize': 20, 'fontweight': 'bold', 'color': 'darkblue'})
plt.xlabel('Price')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```


#### Observations from the Histogram:

1. **Long Tail:**
   - The histogram exhibits a long tail towards the right side, indicating that there are diamonds with relatively higher prices compared to the majority.

2. **Two Peaks:**
   - Notably, there are two peaks in the data. One peak is observed in the range of 500-1000 USD, while the other peak is around 4000-5000 USD.
   - This bimodal distribution suggests the presence of two distinct groups or patterns within the dataset, each with its characteristic pricing range.

By incorporating these observations, we gain a deeper understanding of the distribution patterns and potential groupings within the diamond prices, setting the stage for further analysis.


### Exploring Variation in Diamond Cut Quality

In this analysis, we shift our focus to the quality of cut as a variable of interest. Unlike continuous variables such as price or carat, the variable 'cut' consists of categories, making a histogram unsuitable for analysis. Instead, we will use a bar plot to compare the counts of observations in each category, providing insights into the distribution of diamond cut qualities.

#### Variable: Cut Quality

- **Cut Categories:**
  - The variable 'cut' represents different qualities of cut in diamonds, categorized as 'Fair,' 'Good,' 'Very Good,' 'Premium,' and 'Ideal.'

#### Utilizing Bar Plots for Visualization:

- **Bar Plots:**
  - Bar plots are ideal for visualizing the distribution of categorical variables.
  - In this case, a bar plot will allow us to compare the counts of observations in each cut category, providing insights into the prevalence of different cut qualities within the dataset.

#### Observations and Considerations:

1. **Comparison of Cut Qualities:**
   - The bar plot will enable us to compare the frequencies of each cut quality category.
   - This analysis aims to identify patterns in the distribution of cut qualities and potential areas for further investigation.

2. **Alternative Plot Types:**
   - While other plot types like pie charts or doughnut charts can represent categorical data, they may not offer an intuitive comparison of counts. Bar plots are preferred for this purpose.

By utilizing a bar plot, we seek to uncover patterns and trends in the distribution of diamond cut qualities, contributing to a comprehensive understanding of the dataset.


```python
# Read the data into a dataframe
data = pd.read_csv("diamonds.csv", header=0)

# As a preparatory step, let's drop the 'Unnamed' column.
data = data.drop("Unnamed: 0", axis=1)

# Plotting a bar plot for the quality of cut
plt.figure(figsize=(10, 6))
data['cut'].value_counts().sort_index().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of Diamond Cut Qualities')
plt.xlabel('Cut Quality')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Before plotting, extract the required data by performing necessary data manipulation
groupdata = data.groupby("cut").count()

# Display the first few rows of the grouped data
print(groupdata.head())

# Create lists which hold the necessary data for plotting
x_vals = []
y_vals = []

# Arrange the elements in the list in increasing order of the quality of cut
# This information can be gathered from the data dictionary
for i in [0, 1, 4, 3, 2]: 
    x_vals.append(groupdata.index[i])  # Cut Quality
    y_vals.append(groupdata.iloc[i, 0])  # Count of observations

# Print the lists for verification
print("Cut Quality:", x_vals)
print("Count of Observations:", y_vals)
```

```python
# Plotting a bar plot for the quality of cut
plt.figure(figsize=(10, 6))
plt.bar(x_vals, y_vals, color='skyblue', edgecolor='black')
plt.title('Distribution of Diamond Cut Qualities')
plt.xlabel('Cut Quality')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

##### Observations:

The bar plot representing the distribution of diamond cut qualities reveals the following observation:

- **Increasing Quality of Cut:**
  - As the quality of cut improves, the number of diamonds increases.
  - The distribution shows a positive correlation between the quality of cut and the count of diamonds.
  
This observation suggests that higher-quality cuts are more prevalent in the dataset, reflecting a positive trend in the distribution of diamond cut qualities.


**The distribution of diamond cut qualities can also be visualized using a pie chart**


```python
# Explode the 'Very Good' slice in the pie chart
explode = (0, 0, 0.1, 0, 0)

# Define pastel colors and counter colors
colors = ['lightblue', 'mistyrose', 'lightgreen', 'lightcoral', 'lightyellow']
#counter_colors = ['darkblue', 'darkred', 'darkgreen', 'darkred', 'darkorange']

# Plotting a pie chart for the distribution of diamond cut qualities
plt.figure(figsize=(8, 8))
plt.pie(y_vals, explode=explode, labels=x_vals, autopct='%1.1f%%', startangle=90, colors=colors, pctdistance=0.85)
plt.title("Distribution based on quality of cut\n", fontdict={'fontsize': 20, 'fontweight': 'bold', 'color': 'black'})
plt.show()


```

#### Comparison between Pie Chart and Bar Diagram

While a pie chart may be visually appealing, it may not be the most effective tool for representing information when compared to a bar diagram. The pie chart lacks clarity in showcasing the relative differences between sectors, making it challenging to discern changes accurately. It becomes essential to include written values to enhance understanding.

In contrast, the bar diagram used earlier provides a clearer representation of the distribution, allowing for easy interpretation of the relative differences without the need for data labels. The visual impact of the bar diagram facilitates a more straightforward analysis of the data.



### Investigating Price Variation with Carat

To explore the relationship between diamond prices and carat weight, a scatter plot serves as the ideal visualization. A scatter plot allows for the examination of the correlation between two continuous variables, revealing patterns and trends in the data. In this analysis, we aim to discern any notable variations in diamond prices concerning their carat weight. The scatter plot will provide insights into the potential relationship between these two key attributes.


```python
# Scatter plot for investigating price variation with carat
plt.figure(figsize=(10, 6))
plt.scatter(data['carat'], data['price'], color='purple', alpha=0.5)

# Adding and formatting title and labels
plt.title('Price Variation with Carat', fontdict={'fontsize': 20, 'fontweight': 'bold', 'color': 'black'})
plt.xlabel('Carat', fontsize=16)
plt.ylabel('Price', fontsize=16)

# Show the plot
plt.show()
```

The scatter plot appears crowded due to the abundance of data points. To address this, there are several approaches. One option is to draw a smaller sample.Another approach is to use smaller and more transparent markers, or alternatively, enlarge the plot for better visualization.


```python
# Scatter plot with smaller and more transparent markers
plt.figure(figsize=(12, 8))
plt.scatter(data['carat'], data['price'], color='purple', alpha=0.2, s=10)

# Adding and formatting title
plt.title("Scatter Plot: Variation of Price with Carats\n", fontdict={'fontsize': 20, 'fontweight': 'bold', 'color': 'purple'})

# Adding labels to the axes
plt.xlabel("Carats", fontsize=16, fontweight='bold', color='black')
plt.ylabel("Price", fontsize=16, fontweight='bold', color='black')

# Display the plot
plt.show()


```

Observations:

A noticeable pattern indicates that as the carat weight increases, the corresponding price also tends to rise.
Specific carat weights seem to be more prevalent in the manufacturing of diamonds.
Considerable price variation is evident even among diamonds with identical weights.


#### Investigation of Price Variation with Cut:

1. Since "cut" is a categorical variable, a scatter plot is not suitable for analysis.
2. Explore alternative methods, such as incorporating colors in the plot to represent cut information.
3. Consider dividing the data based on cut and create box plots for each category.
4. Box plots will provide insights into the statistical variation of prices across different cuts.


```python
# Create the required dataframes
data_fair = data[data['cut'] == 'Fair']
data_good = data[data['cut'] == 'Good']
data_vgood = data[data['cut'] == 'Very Good']
data_premium = data[data['cut'] == 'Premium']
data_ideal = data[data['cut'] == 'Ideal']

# Plot box plots for each cut category
plt.boxplot([data_fair['price'], data_good['price'], data_vgood['price'], data_premium['price'], data_ideal['price']],medianprops={'color': 'Green'})
plt.show()

```

##### Outlier Handling:

It's evident that there are numerous outliers in the dataset, complicating the analysis. To streamline our investigation, we've decided to exclude all observations with prices exceeding 12,500 USD.


```python
# Filtering data to exclude outliers
data_fair = data[(data['cut'] == 'Fair') & (data['price'] < 12500)]
data_good = data[(data['cut'] == 'Good') & (data['price'] < 12500)]
data_vgood = data[(data['cut'] == 'Very Good') & (data['price'] < 12500)]
data_premium = data[(data['cut'] == 'Premium') & (data['price'] < 12500)]
data_ideal = data[(data['cut'] == 'Ideal') & (data['price'] < 12500)]

# Plotting boxplots
plt.figure(figsize=[16, 10])
plt.boxplot([data_fair['price'], data_good['price'], data_vgood['price'], data_premium['price'], data_ideal['price']])

# Setting x-axis ticks and labels
plt.xticks((1, 2, 3, 4, 5), ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'])

# Adding and formatting title
plt.title("Price vs Type of Cut\n", fontdict={'fontsize': 20, 'fontweight': 20, 'color': 'black'})

# Labeling Axes
plt.xlabel("Type of Cut", fontdict={'fontsize': 12, 'fontweight': 20, 'color': 'Black'})
plt.ylabel("Price", fontdict={'fontsize': 12, 'fontweight': 20, 'color': 'Black'})

plt.show()
```

**Cautionary Note:**

When making modifications to exclude outliers, it is crucial to consider the potential impact on the overall data. The effect on the averages of different classes may vary, and caution is advised.

**Observations:**

1. The median price of the ideal cut is the least among all the five.
2. There are a significant number of outliers in the ideal cut, even after the modification.
